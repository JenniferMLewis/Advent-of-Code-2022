```python
with open('input.txt') as f:
    lines = f.readlines()
```

```python
lines_test = lines
```

```python
calories = []

for line in lines_test:
    cal = line.strip().strip('[').strip(']').split(',')
    calories.append(cal)
```

```python
elf_cals = pd.DataFrame(calories)
```

```python
elf_cals
```

```python
clean_cals = list(elf_cals[0])
```

```python
elf_try = [list(elf) for k, elf in groupby(clean_cals, key=bool) if k]
```

```python
elf_df = pd.DataFrame(elf_try).T
# elf_df = pd.DataFrame(elf_try)
```

```python
elf_df
```

```python
# elf_df.columns = ["elf1", "elf2","elf3", "elf4", "elf5", "elf6", "elf7", "elf8", "elf9", "elf10", "elf11", "elf12", "elf13", "elf14", "elf15"]
```

```python
elf_df.isnull()
```

```python
elf_df = elf_df.fillna(0)
```

```python
elf_df = elf_df.astype(int)
```

```python
elf_sums = pd.Series(elf_df.sum())
```

```python
elf_sums.max()
```

By the time you calculate the answer to the Elves' question, they've already realized that the Elf carrying the most Calories of food might eventually run out of snacks.

To avoid this unacceptable situation, the Elves would instead like to know the total Calories carried by the top three Elves carrying the most Calories. That way, even if one of those Elves runs out of snacks, they still have two backups.

In the example above, the top three Elves are the fourth Elf (with 24000 Calories), then the third Elf (with 11000 Calories), then the fifth Elf (with 10000 Calories). The sum of the Calories carried by these three elves is 45000.

Find the top three Elves carrying the most Calories. How many Calories are those Elves carrying in total?

```python
first_elf = elf_sums.max()
first_elf
```

```python
elf_sums[elf_sums == first_elf]
```

```python
first_index = elf_sums[elf_sums == first_elf].index
```

```python
first_index
```

```python
elf_sums = elf_sums.drop(first_index)
```

```python
second_elf = elf_sums.max()
second_elf
```

```python
second_index = elf_sums[elf_sums == second_elf].index
```

```python
elf_sums = elf_sums.drop(second_index)
```

```python
third_elf = elf_sums.max()
third_elf
```

```python
first_elf + second_elf + third_elf
```
